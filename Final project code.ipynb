```python
import webbrowser
import string
import speech_recognition as sr
import sys
from gtts import gTTS
import os
import vlc
import time

import sys

try:
    import Tkinter as tk
except ImportError:
    import tkinter as tk

try:
    import ttk
    py3 = False
except ImportError:
    import tkinter.ttk as ttk
    py3 = True

import sanyaProject_support

def vp_start_gui():
    '''Starting point when module is the main routine.'''
    global val, w, root
    root = tk.Tk()
    top = Toplevel1 (root)
    sanyaProject_support.init(root, top)
    root.mainloop()

w = None
def create_Toplevel1(root, *args, **kwargs):
    '''Starting point when module is imported by another program.'''
    global w, w_win, rt
    rt = root
    w = tk.Toplevel (root)
    top = Toplevel1 (w)
    sanyaProject_support.init(w, top, *args, **kwargs)
    return (w, top)

def destroy_Toplevel1():
    global w
    w.destroy()
    w = None

class Toplevel1:
    def recognize_speech_from_mic(recognizer, microphone):
        # check that recognizer and microphone arguments are appropriate type
        if not isinstance(recognizer, sr.Recognizer):
            raise TypeError("`recognizer` must be `Recognizer` instance")

        if not isinstance(microphone, sr.Microphone):
            raise TypeError("`microphone` must be `Microphone` instance")

        # adjust the recognizer sensitivity to ambient noise and record audio
        # from the microphone
        with microphone as source:
            recognizer.adjust_for_ambient_noise(source)#noise removed
            audio = recognizer.listen(source)#listen from mic

        # set up the response object
        response = {
            "success": True,
            "error": None,
            "transcription": None
        }

        # try recognizing the speech in the recording
        # if a RequestError or UnknownValueError exception is caught,
        #     update the response object accordingly
        try:
            response["transcription"] = recognizer.recognize_google(audio) #recognize the input speech (in english)
        except sr.RequestError:
            # API was unreachable or unresponsive
            response["success"] = False
            response["error"] = "API unavailable"
        except sr.UnknownValueError:
            # speech was unintelligible
            response["error"] = "Unable to recognize speech"

        return response
    
    def run(self):
        # obtain audio 
        self.Text2.delete('1.0',tk.END)
        self.Text1.delete('1.0',tk.END)
        NUM_TRIES = 3
        recognizer = sr.Recognizer() #object of class Recognizer()
        microphone = sr.Microphone()
        tts=gTTS(text="Hello, what can i help you find?[Player,Weather,restaurant, or just to search the web: ]",lang='en')
        tts.save('gm.mp3')
        p = vlc.MediaPlayer("gm.mp3")
        p.play()
        #playsound.playsound('gm.mp3')
        time.sleep(5.8)
        print("Hello, what can i help you find?[Player,Weather,restaurant, or just to search the web: ]")
        self.Text2.insert(tk.END,"Hello, what can i help you find?[Player,Weather,restaurant, or just to search the web: ]")
        audio = 0
        guess=0
        for i in range(NUM_TRIES):
            print('Audio {}. Speak!'.format(i+1))
            guess = Toplevel1.recognize_speech_from_mic(recognizer, microphone) # guess will be in form of tuple
            #like {'success': True, 'error': None, 'transcription': 'I want to know about restaurants'}
            if type(guess["transcription"])==str:
                break
            if not guess["success"]:
                continue
        print(guess["transcription"])
        self.Text1.insert(tk.END,str(guess["transcription"]))
        audio=guess["transcription"]
        if(str(audio)=="None"):
            tts=gTTS(text="YOU DID NOT SAY ANYTHING",lang='en')
            tts.save('gm.mp3')
            p = vlc.MediaPlayer("gm.mp3")
            p.play()
            #self.Text2.delete('1.0',tk.END)
            self.Text2.insert(tk.END,"\nYOU DID NOT SAY ANYTHING")
            print("YOU DID NOT SAY ANYTHING")
            return

        if "player" in audio:
            # obtain audio from the microphone
            url = 'https://en.wikipedia.org/wiki/'
            with sr.Microphone() as source:
                tts=gTTS(text="Say any sports players name to see there Bio",lang='en')
                tts.save('gm.mp3')
                p = vlc.MediaPlayer("gm.mp3")
                p.play()
                time.sleep(2.5)
                #self.Text2.delete('1.0',tk.END)
                self.Text2.insert(tk.END,"\nSay any sports players name to see there Bio:")
                print("Say any sports players name to see there Bio: ")
                audio2 = Toplevel1.recognize_speech_from_mic(recognizer, microphone)
                a2=audio2['transcription']
                if(str(a2)=="None"):
                    tts=gTTS(text="YOU DID NOT SAY ANYTHING",lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                    #self.Text2.delete('1.0',tk.END)
                    self.Text2.insert(tk.END,"\nYOU DID NOT SAY ANYTHING")
                    print("YOU DID NOT SAY ANYTHING")
                    return

                try:
                    #self.Text2.delete('1.0',tk.END)
                    self.Text1.insert(tk.END,"\nGoogle Speech Recognition thinks you said " + str(a2))
                    print("Google Speech Recognition thinks you said " + str(a2))
                    tts=gTTS(text="Google Speech Recognition thinks you said. " + str(a2),lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                    webbrowser.open_new(url+str(a2))
                except sr.UnknownValueError:
                    #self.Text2.delete('1.0',tk.END)
                    self.Text2.insert(tk.END,"\nGoogle Speech Recognition could not understand audio")
                    print("Google Speech Recognition could not understand audio")
                    tts=gTTS(text="Google Speech Recognition could not understand audio",lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                except sr.RequestError as e:
                       print("Could not request results from Google Speech Recognition service; {0}".format(e))


        elif "weather" in audio:
            # obtain audio from the microphone
            url3 = 'https://www.timeanddate.com/weather/india/'
            with sr.Microphone() as source:
                tts=gTTS(text="Please say a city",lang='en')
                tts.save('gm.mp3')
                p = vlc.MediaPlayer("gm.mp3")
                p.play()
                time.sleep(1)
                #self.Text2.delete('1.0',tk.END)
                self.Text2.insert(tk.END,"\nPlease say a city: ")
                print("Please say a city: ")
                audio2 = Toplevel1.recognize_speech_from_mic(recognizer, microphone)
                a2=audio2['transcription']
                if(str(a2)=="None"):
                    tts=gTTS(text="YOU DID NOT SAY ANYTHING",lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                    self.Text2.insert(tk.END,"\nYOU DID NOT SAY ANYTHING")
                    print("YOU DID NOT SAY ANYTHING")
                    return

                try:
                    self.Text1.insert(tk.END,"\nGoogle Speech Recognition thinks you said " + str(a2))
                    print("Google Speech Recognition thinks you said " + str(a2))
                    tts=gTTS(text="Google Speech Recognition thinks you said. " + str(a2),lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                    webbrowser.open_new(url3+str(a2))
                except sr.UnknownValueError:
                    self.Text2.insert(tk.END,"\nGoogle Speech Recognition could not understand audio")
                    print("Google Speech Recognition could not understand audio")
                    tts=gTTS(text="Google Speech Recognition could not understand audio",lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                except sr.RequestError as e:
                       print("Could not request results from Google Speech Recognition service; {0}".format(e))


        elif "restaurant" in audio:
            # obtain audio from the microphone
            url4 = 'https://www.zomato.com/'
            with sr.Microphone() as source:
                tts=gTTS(text="Please state a location to search for restaurant",lang='en')
                tts.save('gm.mp3')
                p = vlc.MediaPlayer("gm.mp3")
                p.play()
                time.sleep(2.7)
                self.Text2.insert(tk.END,"\nPlease state a location to search for restaurant: ")
                print("Please state a location to search for restaurant: ")
                audio2 = Toplevel1.recognize_speech_from_mic(recognizer, microphone)
                a2=audio2['transcription']
                if(str(a2)=="None"):
                    tts=gTTS(text="YOU DID NOT SAY ANYTHING",lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                    self.Text2.insert(tk.END,"\nYOU DID NOT SAY ANYTHING")
                    print("YOU DID NOT SAY ANYTHING")
                    return
                try:
                    self.Text1.insert(tk.END,"\nGoogle Speech Recognition thinks you said " +str(a2))
                    print("Google Speech Recognition thinks you said " +str(a2))
                    tts=gTTS(text="Google Speech Recognition thinks you said. " + str(a2),lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                    webbrowser.open_new(url4+str(a2))
                except sr.UnknownValueError:
                    self.Text2.insert(tk.END,"\nGoogle Speech Recognition could not understand audio")
                    print("Google Speech Recognition could not understand audio")
                    tts=gTTS(text="Google Speech Recognition could not understand audio",lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                except sr.RequestError as e:
                       print("Could not request results from Google Speech Recognition service; {0}".format(e))

        elif "web" in audio:
            # obtain audio from the microphone

            with sr.Microphone() as source:
                tts=gTTS(text="what website would you like me to search for you?",lang='en')
                tts.save('gm.mp3')
                p = vlc.MediaPlayer("gm.mp3")
                p.play()
                time.sleep(2.7)
                self.Text2.insert(tk.END,"\nwhat website would you like me to search for you? ")
                print("what website would you like me to search for you? ")
                audio2 = Toplevel1.recognize_speech_from_mic(recognizer, microphone)
                a2=audio2['transcription']
                if(str(a2)=="None"):
                    tts=gTTS(text="YOU DID NOT SAY ANYTHING",lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                    self.Text2.insert(tk.END,"\nYOU DID NOT SAY ANYTHING")
                    print("YOU DID NOT SAY ANYTHING")
                    return

                try:
                    self.Text1.insert(tk.END,"\nGoogle Speech Recognition thinks you said " + str(a2))
                    print("Google Speech Recognition thinks you said " + str(a2))
                    tts=gTTS(text="Google Speech Recognition thinks you said. " + str(a2),lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                    webbrowser.open_new("https://"+str(a2)+".com")
                except sr.UnknownValueError:
                    self.Text2.insert(tk.END,"\nGoogle Speech Recognition could not understand audio")
                    print("Google Speech Recognition could not understand audio")
                    tts=gTTS(text="Google Speech Recognition could not understand audio",lang='en')
                    tts.save('gm.mp3')
                    p = vlc.MediaPlayer("gm.mp3")
                    p.play()
                except sr.RequestError as e:
                       print("Could not request results from Google Speech Recognition service; {0}".format(e)) 
        else:
            tts=gTTS(text=audio+" Not recognized",lang='en')
            tts.save('gm.mp3')
            p = vlc.MediaPlayer("gm.mp3")
            p.play()
            print(audio," Not recognized")


    def __init__(self, top=None):
        '''This class configures and populates the toplevel window.
           top is the toplevel containing window.'''
        _bgcolor = '#d9d9d9'  # X11 color: 'gray85'
        _fgcolor = '#000000'  # X11 color: 'black'
        _compcolor = '#d9d9d9' # X11 color: 'gray85'
        _ana1color = '#d9d9d9' # X11 color: 'gray85' 
        _ana2color = '#d9d9d9' # X11 color: 'gray85' 

        top.geometry("600x450+650+150")
        top.title("Voice Recognition")
        top.configure(background="#ffffff")

        self.Frame1 = tk.Frame(top)
        self.Frame1.place(relx=0.017, rely=0.022, relheight=0.944
                , relwidth=0.958)
        self.Frame1.configure(relief='groove')
        self.Frame1.configure(borderwidth="2")
        self.Frame1.configure(relief='groove')
        self.Frame1.configure(background="#969696")
        self.Frame1.configure(width=575)

        self.Frame2 = tk.Frame(self.Frame1)
        self.Frame2.place(relx=0.087, rely=0.118, relheight=0.482
                , relwidth=0.809)
        self.Frame2.configure(relief='groove')
        self.Frame2.configure(borderwidth="2")
        self.Frame2.configure(relief='groove')
        self.Frame2.configure(background="#505a5b")
        self.Frame2.configure(width=465)

        self.Label1 = tk.Label(self.Frame2)
        self.Label1.place(relx=0.022, rely=0.098, height=26, width=62)
        self.Label1.configure(background="#505a5b")
        self.Label1.configure(disabledforeground="#a3a3a3")
        self.Label1.configure(foreground="#000000")
        self.Label1.configure(text='''Input:''')
        self.Label1.configure(width=62)

        self.Label2 = tk.Label(self.Frame2)
        self.Label2.place(relx=0.022, rely=0.537, height=26, width=55)
        self.Label2.configure(background="#505a5b")
        self.Label2.configure(disabledforeground="#a3a3a3")
        self.Label2.configure(foreground="#000000")
        self.Label2.configure(text='''Output:''')

        self.Text1 = tk.Text(self.Frame2)
        self.Text1.place(relx=0.151, rely=0.098, relheight=0.41, relwidth=0.804)
        self.Text1.configure(background="white")
        self.Text1.configure(font="TkTextFont")
        self.Text1.configure(foreground="black")
        self.Text1.configure(highlightbackground="#d9d9d9")
        self.Text1.configure(highlightcolor="black")
        self.Text1.configure(insertbackground="black")
        self.Text1.configure(selectbackground="#c4c4c4")
        self.Text1.configure(selectforeground="black")
        self.Text1.configure(width=374)
        self.Text1.configure(wrap='word')

        self.Text2 = tk.Text(self.Frame2)
        self.Text2.place(relx=0.151, rely=0.537, relheight=0.41, relwidth=0.804)
        self.Text2.configure(background="white")
        self.Text2.configure(font="TkTextFont")
        self.Text2.configure(foreground="black")
        self.Text2.configure(highlightbackground="#d9d9d9")
        self.Text2.configure(highlightcolor="black")
        self.Text2.configure(insertbackground="black")
        self.Text2.configure(selectbackground="#c4c4c4")
        self.Text2.configure(selectforeground="black")
        self.Text2.configure(width=374)
        self.Text2.configure(wrap='word')

        self.Button1 = tk.Button(self.Frame1)
        self.Button1.place(relx=0.33, rely=0.682, height=83, width=226)
        self.Button1.configure(activebackground="#d9d9d9")
        self.Button1.configure(activeforeground="#000000")
        self.Button1.configure(background="#ffffff")
        self.Button1.configure(disabledforeground="#a3a3a3")
        self.Button1.configure(foreground="#000000")
        self.Button1.configure(highlightbackground="#d9d9d9")
        self.Button1.configure(highlightcolor="black")
        self.Button1.configure(pady="0")
        self.Button1.configure(text='''Start Mic''')
        self.Button1.configure(width=226)
        self.Button1.configure(command=self.run)

if __name__ == '__main__':
    vp_start_gui()

```
